### float 타입 컬럼들의 이상치 확인
- 적절한 이상치 처리 (교체)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 한글 폰트 설정 (예: 맑은 고딕)
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# Boxplot 시각화
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y=df[df.columns[5]])
plt.title('학점 이상값')
plt.subplot(1, 2, 2)
sns.boxplot(y=df[df.columns[7]])
plt.title('공부 시간 이상값')
plt.show()
```

```python
import numpy as np

def replace_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # 이상값을 상한선 또는 하한선으로 대체
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

    return df


df = replace_outliers(df, df.columns[5])
df = replace_outliers(df, df.columns[7])

# Boxplot 시각화
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y=df[df.columns[5]])
plt.title('학점 이상값')
plt.subplot(1, 2, 2)
sns.boxplot(y=df[df.columns[7]])
plt.title('공부 시간 이상값')
plt.show()
```

### 데이터 학습
- 훈련, 테스트 데이터셋 분류
- 랜덤포레스트 모델 사용
- 클래스 불균형으로 인해 모델 파라미터에 class_weight='balanced' 추가

```python
X = df.drop(columns=[df.columns[0]])
y = df[df.columns[0]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, classification_report, confusion_matrix

model = RandomForestClassifier(
    class_weight='balanced',
    max_depth=6,                # 트리의 최대 깊이
    min_samples_split=3,        # 노드를 분할하기 위한 최소 샘플 수
    min_samples_leaf=6,         # 리프 노드가 되기 위한 최소 샘플 수
    random_state=42
)

# 모델 학습
model.fit(X_train, y_train)

# 예측 및 평가 지표 확인
y_pred = model.predict(X_test)
print(f"accuracy: {accuracy_score(y_test, y_pred):.2f}")
print(f'f1: {f1_score(y_test, y_pred):.2f}')
print(f'precision: {precision_score(y_test, y_pred):.2f}')
print(f'recall: {recall_score(y_test, y_pred):.2f}')
print(f'roc-auc: {roc_auc_score(y_test, model.predict_proba(X_test)[:,1]):.2f}')
```

### 시각화
- 혼동 행렬
- 변수 중요도
- 트리
- 카운트 플롯

```python
print(classification_report(y_test, y_pred))

# 혼동 행렬 시각화
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
# plt.savefig("confusion_matrix.jpg", format='jpg', bbox_inches='tight')
# plt.close()
```

```python
# 변수 중요도 추출
importance = model.feature_importances_

# 중요도 순서에 따라 변수 이름 정렬
indices = np.argsort(importance)[::-1]
feature_names = X.columns[indices]

# 변수 중요도 시각화
plt.figure(figsize=(12, 8))
plt.title("Feature Importances")
plt.bar(range(X.shape[1]), importance[indices], align="center")
plt.xticks(range(X.shape[1]), feature_names, rotation=45)
plt.xlim([-1, X.shape[1]])
plt.show()
```

```python
from sklearn.tree import export_graphviz, plot_tree
import graphviz
import random

# 랜덤 포레스트에서 idx 번째 트리를 선택하여 시각화
idx = random.randrange(100)
print(f'ex) {idx}번째 트리')
estimator = model.estimators_[idx]

# 트리를 시각화 (텍스트 기반)
plt.figure(figsize=(20, 10))
plot_tree(estimator, filled=True, feature_names=X.columns, rounded=True)
plt.show()

# 트리를 graphviz를 사용해 시각화 (그래픽 기반)
dot_data = export_graphviz(estimator, out_file=None, 
                           feature_names=X.columns,  
                           class_names=['Class 0', 'Class 1'], 
                           filled=True, rounded=True,  
                           special_characters=True)  
graph = graphviz.Source(dot_data)  
graph.render("tree")  # 트리의 PDF 파일을 생성
graph.view()  # 트리의 시각화 확인
```
